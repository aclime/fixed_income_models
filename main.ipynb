```python
import plotly.graph_objects as go
import ipywidgets as widgets
import numpy as np

from models.vasicek import (common_terms,
                            continuous_yc,
                            discrete_yc,
                            )
```

```python
#Initial params
r=0.03      #Short-term interest rate
theta=0.05  #Long-term interest rate
kappa=0.50	#Mean-reversion
sigma =0.02	#Volatility
T=30        #maturity upper bound
T_line=np.arange(1,T+1)



ru=r+kappa*(theta-r)+sigma
rd=r+kappa*(theta-r)-sigma
B,A = common_terms(kappa,T_line,theta,sigma)
ycc=continuous_yc(A,B,r,T_line)
#ycc.shape
dcc=discrete_yc(A,B,r,sigma,kappa,theta,T_line)
#dcc.shape
fig = go.FigureWidget(data=[go.Scatter(x=T_line, y=ycc,name='continuous'),
                            go.Scatter(x=T_line[1:], y=dcc,name='discrete') ])
fig.update_layout(yaxis_tickformat='.2%') 
fig.update_layout(title='Vasicek Model')
#fig.show()

#Param widgets
r_slider = widgets.FloatSlider(min=0.01, max=0.1, step=0.01, value=0.03, description='𝑟:' )
theta_slider = widgets.FloatSlider(min=0.01, max=0.1, step=0.01, value=0.05, description='θ:' )
kappa_slider = widgets.FloatSlider(min=0.01, max=1.0, step=0.25, value=0.5, description='κ:' )
sigma_slider = widgets.FloatSlider(min=0.005, max=0.04, step=0.005, value=0.02, description='σ:' )
ru_display = widgets.Text(value=f'{round(ru*100,2)}%',description='rᵤ',disabled=True)
rd_display = widgets.Text(value=f'{round(rd*100,2)}%',description='rᵈ',disabled=True)
reset_button = widgets.Button(description='reset',disabled=False,button_style='',tooltip='Click me')



#Update functions
def update_r(change):
    r_new = change.new
    sigma,theta,kappa=sigma_slider.value,theta_slider.value,kappa_slider.value
    B,A = common_terms(kappa,T_line,theta,sigma)
    continuous=continuous_yc(A,B,r_new,T_line)
    discrete=discrete_yc(A,B,r_new,sigma,T_line)
    #"""
    ru_new=r_new+kappa*(theta-r_new)+sigma
    rd_new=r_new+kappa*(theta-r_new)-sigma
    ru_display.value=f'{round(ru_new*100,2)}%'
    rd_display.value=f'{round(rd_new*100,2)}%'
    #"""
    with fig.batch_update():
        fig.data[0].y = continuous
        fig.data[1].y = discrete

def update_theta(change):
    theta_new = change.new
    sigma,r,kappa=sigma_slider.value,r_slider.value,kappa_slider.value
    B,A = common_terms(kappa,T_line,theta_new,sigma)
    continuous=continuous_yc(A,B,r,T_line)
    discrete=discrete_yc(A,B,r,sigma,kappa,theta,T_line)
    #"""
    ru_new=r+kappa*(theta_new-r)+sigma
    rd_new=r+kappa*(theta_new-r)-sigma
    ru_display.value=f'{round(ru_new*100,2)}%'
    rd_display.value=f'{round(rd_new*100,2)}%'
    #"""
    with fig.batch_update():
        fig.data[0].y = continuous
        fig.data[1].y = discrete

def update_kappa(change):
    kappa_new = change.new
    sigma,r,theta=sigma_slider.value,r_slider.value,theta_slider.value
    B,A = common_terms(kappa_new,T_line,theta,sigma)
    continuous=continuous_yc(A,B,r,T_line)
    discrete=discrete_yc(A,B,r,sigma,kappa,theta,T_line)
    #"""
    ru_new=r+kappa_new*(theta-r)+sigma
    rd_new=r+kappa_new*(theta-r)-sigma
    ru_display.value=f'{round(ru_new*100,2)}%'
    rd_display.value=f'{round(rd_new*100,2)}%'
    #"""
    with fig.batch_update():
        fig.data[0].y = continuous
        fig.data[1].y = discrete

def update_sigma(change):
    sigma_new = change.new
    kappa,r,theta=kappa_slider.value,r_slider.value,theta_slider.value
    B,A = common_terms(kappa,T_line,theta,sigma_new)
    continuous=continuous_yc(A,B,r,T_line)
    discrete=discrete_yc(A,B,r,sigma_new)
    #"""
    ru_new=r+kappa*(theta-r)+sigma_new
    rd_new=r+kappa*(theta-r)-sigma_new
    ru_display.value=f'{round(ru_new*100,2)}%'
    rd_display.value=f'{round(rd_new*100,2)}%'
    #"""
    with fig.batch_update():
        fig.data[0].y = continuous
        fig.data[1].y = discrete

def reset_params(change):
    kappa_slider.value = kappa
    r_slider.value = r
    theta_slider.value = theta
    sigma_slider.value = sigma
    B,A = common_terms(kappa,T_line,theta,sigma)
    continuous=continuous_yc(A,B,r,T_line)
    discrete=discrete_yc(A,B,r,sigma,kappa,theta,T_line)
    ru_new=r+kappa*(theta-r)+sigma
    rd_new=r+kappa*(theta-r)-sigma
    ru_display.value=f'{round(ru_new*100,2)}%'
    rd_display.value=f'{round(rd_new*100,2)}%'
    with fig.batch_update():
        fig.data[0].y = continuous
        fig.data[1].y = discrete



#Observers
r_slider.observe(update_r, names='value')
theta_slider.observe(update_theta, names='value')
kappa_slider.observe(update_kappa, names='value')
sigma_slider.observe(update_sigma, names='value')

reset_button.on_click(reset_params)



display( widgets.VBox([ r_slider,theta_slider,kappa_slider,sigma_slider,
                       widgets.HBox([ru_display,rd_display]),
                       reset_button,
                        fig])  )
```
